```python
import numpy as np
import jax
import jax.numpy as jnp
```

```python
f = lambda x: (x - 3)**2 + 1

grad_f = jax.grad(f)
grad_grad_f = jax.grad(grad_f)

grad_f(3.0)
```

```python
grad_grad_f(3.0)
```

```python
g = lambda x: x[0]**2 + 2*x[0]*x[1] + 2*x[1]**2

grad_g = jax.grad(g)

hess_g = jax.hessian(g)
```

```python
grad_g(jnp.array([0.0, 0.0]))
```

```python
hessian_at_0 = hess_g(jnp.array([0.0, 0.0]))
```

```python
jnp.linalg.eigvals(hessian_at_0)
```

```python
linear_form = lambda x, a: x @ a

grad_linear_form = jax.grad(linear_form, argnums=0)

key_a = jax.random.key(0)
key_x = jax.random.key(1)

a = jax.random.normal(key_a, (5,))
x = jax.random.normal(key_x, (5,))

print(x)
print(a)
grad_linear_form(x, a)
```

```python
quad_form = lambda x, A: x.T @ A @ x

grad_quad_form = jax.grad(quad_form, argnums=0)

key_a = jax.random.key(0)
key_x = jax.random.key(1)

A = jax.random.normal(key_a, (5, 5))
x = jax.random.normal(key_x, (5,))

grad_quad_form(x, A)
```

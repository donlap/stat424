```python
import numpy as np
import cvxpy as cp
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X, y = data.data, data.target
feature_names = data.feature_names
```

```python
X
```

```python
feature_names
```

```python
X_scaled = (X - X.mean()) / X.std()

```

```python
X_final = np.hstack([np.ones((X_scaled.shape[0], 1)), X_scaled])
X_final
```

```python
d = X_final.shape[1]
lamb = 10.0

beta = cp.Variable(d)
loss = cp.sum_squares(X_final @ beta - y) / (len(y))
reg1 = lamb * cp.norm1(beta[1:])

objective = cp.Minimize(loss + reg1)
prob = cp.Problem(objective)
prob.solve()
```
